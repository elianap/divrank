```python
import pickle
import pandas as pd
import os
```

# DivRank results

```python
name_output_dir = 'results/divrank_results_mitigation'

for dataset_name in ['law_school', 'compas-poc', 'german', 'artificial_1']:

        input_dir = os.path.join(name_output_dir, dataset_name)
        with open(os.path.join(input_dir, f'{dataset_name}_divrank_results_mitigation.pickle'), 'rb') as fp:
                ranking_results = pickle.load( fp)


        from utils_evaluation import print_table_ranking_res

        r = pd.DataFrame(ranking_results).T
        print(dataset_name)
        r_print = print_table_ranking_res(r)
        display(r_print)

        with open(os.path.join(input_dir, f'{dataset_name}_divrank_results_mitigation_time.pickle'), 'rb') as fp:

                divrank_time = pickle.load( fp)
        df_divrank_time = pd.DataFrame.from_dict(divrank_time, orient='index', columns = ['time'])
        display(df_divrank_time)

        print(f"Max for {dataset_name}:  {df_divrank_time['time'].max():.2f}\n")
```

# Feldman results

```python
name_output_dir = 'results/feldman_results_mitigation'

for dataset_name in ['law_school', 'compas-poc', 'german', 'artificial_1']:

        input_dir = os.path.join(name_output_dir, dataset_name)
        if os.path.isfile(os.path.join(input_dir, f'{dataset_name}_feldman_results_mitigation.pickle')):
                with open(os.path.join(input_dir, f'{dataset_name}_feldman_results_mitigation.pickle'), 'rb') as fp:
                        ranking_results = pickle.load( fp)


                from utils_evaluation import print_table_ranking_res

                r = pd.DataFrame(ranking_results).T
                print(dataset_name)
                r_print = print_table_ranking_res(r)
                display(r_print)

                with open(os.path.join(input_dir, f'{dataset_name}_feldman_results_mitigation_time.pickle'), 'rb') as fp:

                        res_time = pickle.load( fp)
                df_res_time = pd.DataFrame.from_dict(res_time, orient='index', columns = ['time'])
                display(df_res_time)
                print(f"Max for {dataset_name}:  {df_res_time['time'].max():.2f}\n")
        else:
                print(f'{dataset_name} not available')
```

# CFA

```python
name_output_dir = 'results/cfa_results_mitigation'

for dataset_name in ['law_school', 'compas-poc', 'german', 'artificial_1']:

        input_dir = os.path.join(name_output_dir, dataset_name)

        if os.path.isfile(os.path.join(input_dir, f'{dataset_name}_cfa_results.pickle')):
            with open(os.path.join(input_dir, f'{dataset_name}_cfa_results.pickle'), 'rb') as fp:
                    ranking_results = pickle.load( fp)


            from utils_evaluation import print_table_ranking_res

            r = pd.DataFrame(ranking_results).T
            print(dataset_name)
            r_print = print_table_ranking_res(r)
            display(r_print)


            with open(os.path.join(input_dir, f'{dataset_name}_cfa_time.pickle'), 'rb') as fp:

                    res_time = pickle.load( fp)
            df_res_time = pd.DataFrame.from_dict(res_time, orient='index', columns = ['time'])
            display(df_res_time)
            print(f"Max for {dataset_name}:  {df_res_time['time'].max():.2f} s\n   - {df_res_time['time'].max()/60:.2f} m")

        else:
            print(f'{dataset_name} NOT AVAILABLE')
```
